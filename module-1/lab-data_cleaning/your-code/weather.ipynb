## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd
weatherdf = pd.read_csv('../weather-raw.csv')
display(weatherdf)
```

```python
#borro el id y el año porque se repite y no me hace falta ya que sé que son datos del 2010
weatherdf = weatherdf.drop(columns=["id","year"])
```

```python
oddweatherdf = weatherdf.iloc[1::2]
evenweatherdf = weatherdf.iloc[::2]
display(oddweatherdf)
display(evenweatherdf)
```

```python
evenweatherdf["mediamax"]= evenweatherdf.loc[:, evenweatherdf.columns != 'month'].mean(axis=1,skipna=True)
evenweatherdf["minMax"]= evenweatherdf.loc[:, evenweatherdf.columns != 'month'].min(axis=1,skipna=True)
evenweatherdf["maxMax"]= evenweatherdf.loc[:, evenweatherdf.columns != 'month'].max(axis=1,skipna=True)
oddweatherdf["mediamin"]= oddweatherdf.loc[:, oddweatherdf.columns != 'month'].mean(axis=1,skipna=True)
oddweatherdf["minMin"]= oddweatherdf.loc[:, oddweatherdf.columns != 'month'].min(axis=1,skipna=True)
oddweatherdf["maxMin"]= oddweatherdf.loc[:, oddweatherdf.columns != 'month'].max(axis=1,skipna=True)
```

```python
display(evenweatherdf)
display(oddweatherdf)
```

```python
evenweatherdf_clean = evenweatherdf[["month","element","mediamax","minMax","maxMax"]]
oddweatherdf_clean = oddweatherdf[["month","element","mediamin","minMin","maxMin"]]
mergedweather = evenweatherdf_clean.merge(oddweatherdf_clean, on="month")
display(mergedweather)
```
